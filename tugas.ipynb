```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection 
from matplotlib import style
```

x - 2y + z = 6

3x + y -2z = 4

7x - 6y - z = 10



### **SPLTV (Sistem Persamaan Linear Tiga Variabel)**

```python
a = np.array([[1,-2,1],[3,1,-2],[7,-6,-1]])
b = np.array([6,4,10])
result = np.linalg.solve(a,b)

print(f'Nilai X = {round(result[0])} \nNilai Y = {round(result[1])} \nNilai Z = {round(result[2])}')
```

## **3D Plot SPLTV**

```python
plt.figure('SPLTV',figsize=(15,7))
ax1 = plt.subplot(121, projection='3d')

#x-2y+z=6

x1=np.array([6, 0, 0])
y1=np.array([0, -3, 0])
z1=np.array([0, 0, 6]) 
ax1.scatter(x1, y1, z1, s=110)

# 1. MEMBUAT LIST TITIK KOORDINAT
titik1 = [list(zip(x1, y1, z1))]

# # # 2. MEMBUAT 3d polygons DAM PARAMETER
polygon1 = Poly3DCollection(titik1, alpha=.55, facecolor='r')

# 3. MEMBUAT SEGITIGA
plt.gca().add_collection3d(polygon1)

plt.title('x - 2y + z = 6')
ax1.set_xlabel('Nilai X')
ax1.set_ylabel('Nilai Y')
ax1.set_zlabel('Nilai Z')
plt.show()
```

```python
plt.figure('SPLTV',figsize=(15,7))
ax2 = plt.subplot(121,projection='3d')

#3x + y - 2z = 4
x1=np.array([4/3, 0, 0])
y1=np.array([0, 4, 0])
z1=np.array([0, 0, -2]) 

ax2.scatter(x1,y1,z1, s=110)
titik2 = [list(zip(x1, y1, z1))]
polygon2 = Poly3DCollection(titik2, alpha=.55, facecolor='g')
plt.gca().add_collection3d(polygon2)

plt.title('3x + y - 2z = 4')
ax2.set_xlabel('Nilai X')
ax2.set_ylabel('Nilai Y')
ax2.set_zlabel('Nilai Z')

plt.show()
```
